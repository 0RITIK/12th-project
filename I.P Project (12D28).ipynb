```python
def visuals():
    opt=""
    print()
    print("^" * 20 , "Data Visualization Menu" , "^" * 20)
    print('1.Line Chart\n')
    print('2.Bar Chart\n')
    print('3.Exit\n')
    opt = int(input('Enter your choice:'))
    if opt==1:
        line()
    if opt==2:
        bar()
    if opt==3:
        a = input('Press m For Main Menu\nPress e For Exit\n')
    if a == 'm':
        menu()
    if a == 'e':
        print('Thanks!!')
        exit()
```

```python
def line():
    line = pd.read_csv('Project.csv')
    df = pd.DataFrame(line)
    x = (df['Year'])
    plt.plot(x,df['0-24 (%)'],color='#FFC000',marker='*',linestyle='-.',linewidth=2)
    plt.plot(x,df['25-44 (%)'],color='cyan',marker=',',linestyle=':',linewidth=3)
    plt.plot(x,df['45-64 (%)'],color='green',marker='1',linestyle='--',linewidth=2)
    plt.plot(x,df['65&Above (%)'],color='blue',marker='x',linestyle='--',linewidth=2)
    plt.plot(x,df['Unreported (%)'],color='black',linewidth=3,marker='x',linestyle=':',)
    plt.legend([['0-24'],['25-44'],['45-64'],['65&Above'],['Unreported']])
    plt.grid()
    plt.xlabel('Years --->')
    plt.ylabel('Arrivals(%)--->')
    plt.title('FTAs IN INDIA IN 2006-2020 ACCORDING TO AGE GROUP in (%)')
    plt.show()
    menu()
```

```python
def bar():
    c=pd.read_csv('Project.csv')
    bar = pd.DataFrame(c)
    bar=bar.set_index('Year')
    df = pd.DataFrame(bar)
    df.plot(kind='bar',stacked=True)
    plt.xlabel('Years')
    plt.ylabel('FTAs(%) according to age group [Total 100%] -->')
    plt.title('FTAs IN INDIA IN 2006-2020 ACCORDING TO AGE GROUP in (%)')
    plt.show()
    menu()
```
